```python
search_param = {
    "query":{
        "match_all":{}
    }
}

# Initialize the scroll
scroll = es.search(
    index='smb_data',
    scroll='2m',  # Keep the scroll context alive for 2 minutes
    size=1000,  # Number of results to retrieve per scroll
    body=search_param
)

# Get the scroll ID
scroll_id = scroll['_scroll_id']

# Keep track of the progress
all_hits = scroll['hits']['hits']  

# Start scrolling
while True:
    # Get the next batch of results
    response = es.scroll(
        scroll_id=scroll_id,
        scroll='2m'  # Extend the scroll context for another 2 minutes
    )

    # Break the loop if there are no more results
    if not response['hits']['hits']:
        break

    # Update the scroll ID
    scroll_id = response['_scroll_id']

    # Add the new results, all_hits is a list
    all_hits.extend(response['hits']['hits'])

# At this point, `all_hits` contains all the results from your query
# Remember to clear the scroll when you're done
es.clear_scroll(scroll_id=scroll_id)


df_from_ES = pd.DataFrame([i['_source'] for i in all_hits])
df_from_ES
```
